```python
from urllib.request import urlopen
from urllib.parse import urlencode, unquote, quote_plus
import urllib
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
# from selenium import webdriver
import time
import datetime
```

```python
weather_file = pd.read_csv('./weather_crawling.csv')
# weather_file
```

### 시작

```python
def enter_weather(weather):
    header = {'User-Agent': 'Mozilla/5.0'} 
    base_url = 'https://www.melon.com/dj/djfinder/djfinder_inform.htm'
    params = {
        'djSearchKeyword':weather
    }
    resp = requests.get(base_url,headers=header,params=params)
    soup = BeautifulSoup(resp.text)
    ply_list_dj = soup.select('div.service_list_play.d_djcol_list li')
    seq = []
    for li in ply_list_dj:
        a_tag = li.find('a')
        par = a_tag.get('href')
        seq.append(int(par[-12:-3]))
    return seq
```

```python
def title_and_singer(seq,info):
    titles = []       
    singers = []
    song = []
    likes = []
    for i in range(len(seq)):
        new_url = 'https://www.melon.com/mymusic/dj/mymusicdjplaylistview_inform.htm'
        new_params = {
        'plylstSeq': seq[i]
        }
        hdr={'User-Agent': 'Mozilla/5.0'}
        new_resp = requests.get(new_url,headers=hdr,params=new_params)
        new_soup = BeautifulSoup(new_resp.text)
        
        song_list = new_soup.find_all('div',class_='wrap_song_info')
        for wrap in song_list:
            title = wrap.find('div',class_='ellipsis rank01')
            singer = wrap.find('div',class_='ellipsis rank02')
            if singer != None:
                leng = round(len(singer.text)/2)
                singers.append(singer.text[:leng].strip())
            if title != None:
                titles.append(title.text.strip())

        for ts in range(len(titles)):
            song.append((titles[ts],singers[ts]))
        likes = likes_num(seq[i])
        for tsl in range(len(likes)):
            if song[tsl] in info:
                info[song[tsl]] = likes[tsl]+100
            else:
                info[song[tsl]] = likes[tsl]
    return info
```

```python
def likes_num(sq):
    likes = []
    new_url = 'https://www.melon.com/mymusic/dj/mymusicdjplaylistview_inform.htm'
    new_params = {
        'plylstSeq': sq
    }                                      
    header = {'User-Agent': 'Mozilla/5.0'}
    new_resp = requests.get(new_url,headers = header,params=new_params)
    new_soup = BeautifulSoup(new_resp.text)
    song_in_plylist = new_soup.select('div.wrap button')

    song_no = []
    for button in song_in_plylist:
        no = button.get('data-song-no')
        if no != None:
            song_no.append(no)
    song_no_select = song_no[0]
    for s in range(1,len(song_no)):
        song_no_select = song_no_select +','+str(song_no[s])
    ply_url = 'https://www.melon.com/commonlike/getSongLike.json'
    ply_params = {
        "contsIds": song_no_select
        }
    referer = 'https://www.melon.com/mymusic/dj/mymusicdjplaylistview_inform.htm?plylstSeq='+str(sq)

    hd = {
        'User-Agent': 'Mozilla/5.0',
        'referer':referer
        }
    ply_resp = requests.get(ply_url,headers=hd,params=ply_params)
    result= json.loads(ply_resp.text)
    extract = result['contsLike']

    for i in range(len(song_no)):
        likes.append(extract[i]['SUMMCNT'])
    return likes
```

```python
def top5_music():
    for k in range(len(weather_file['날씨'])-1):
        info = {}
        date = weather_file['날짜'][k]
        curr_seq = enter_weather(weather_file['날씨'][k])
        result_info = title_and_singer(curr_seq,info)
        result_info = sorted(result_info.items(), key=lambda x: x[1], reverse=True)
        for j in range(5):
            music_result['날짜'].append(date),
            music_result['노래'].append(result_info[j][0][0]),
            music_result['가수'].append(result_info[j][0][1]),
            music_result['좋아요 수'].append(result_info[j][1])
            
    df_music=pd.DataFrame(music_result)
    
    return df_music
```

```python
def append_today_music(data):
    info = {}
    date, music, atist,likeit = [],[],[],[]
    curr_date = weather_file.loc[len(weather_file['날씨'])-1]['날짜']
    curr_weather = weather_file.loc[len(weather_file['날씨'])-1]['날씨']
    curr_seq = enter_weather(curr_weather)
    result_info = title_and_singer(curr_seq,info)
    result_info = sorted(result_info.items(), key=lambda x: x[1], reverse=True)
    
    for j in range(5):
        date.append(curr_date),
        music.append(result_info[j][0][0]),
        atist.append(result_info[j][0][1]),
        likeit.append(result_info[j][1])
    
    
    for k in range(5):
        new_music_data=pd.DataFrame({'날짜':[date[k]],
                 '노래':[music[k]],
                 '가수':[atist[k]],
                 '좋아요 수':[likeit[k]]
                              })

        data = data.append(new_music_data,ignore_index=True)
    
    return data
```

```python
music_result={
        '날짜':[],
        '노래':[],
        '가수':[],
        '좋아요 수':[]
    }
music_output = top5_music()
m = append_today_music(music_output)
m.to_csv('melon_crawling.csv',encoding='utf-8', index=False)
print('=====END Melon Crawling=====')
```
